#### Обучение модели

```python
# для обучения модели будем применять поиск по сетке. Модель возьму - случайный лес. 
#Хотя я прочитала в книге, что в промышленных решениях применяют RandomSearch.
#Но Евгений рекомендовал именно GridSearch. 

from sklearn.model_selection import GridSearchCV

model_forest = RandomForestClassifier(random_state=12345)
params_forest = {'max_depth':range(1,15), 'n_estimators':range(10, 150,20)}
grid_model = GridSearchCV(model_forest, params_forest, cv=5, scoring = ['f1', 'roc_auc'], verbose=True, refit='f1') #verbose показывает время

grid_model.fit(features_train, target_train)
forest_f1_nonb=grid_model.best_score_

```

```python
print ("Метрика F1 на учебной выборке=", grid_model.best_score_)

results = pd.DataFrame(grid_model.cv_results_)
print('ROC AUC', results[results['mean_test_f1']== grid_model.best_score_]['mean_test_roc_auc'])
```

```python
model=grid_model.best_estimator_
model.fit(features_trainplus, target_trainplus)
preds = model.predict(features_test)
#print ("Метрика F1 на тестовой выборке=",f1_score(target_test, preds))

```

```python
# Попробуем другой метод поиска гиперпараметров
from sklearn.model_selection import RandomizedSearchCV

grid_model = RandomizedSearchCV(model_forest, params_forest, cv=5, n_jobs=-1, scoring = ['f1', 'roc_auc'], verbose=True, refit='f1')
grid_model.fit(features_trainplus, target_trainplus)
```

```python
#grid_model.best_estimator_
print ("Метрика F1 на учебной выборке=", grid_model.best_score_)

results = pd.DataFrame(grid_model.cv_results_)
print('ROC AUC', results[results['mean_test_f1']== grid_model.best_score_]['mean_test_roc_auc'])
```

```python
#оценим F1 также на логистической регресии

params_logreg = {'C':range(1,5), 'intercept_scaling':range(1,5)}
model_logreg = LogisticRegression(solver='liblinear', random_state=12345)
grid_model = GridSearchCV(model_logreg, params_logreg, cv=5, scoring = ['f1', 'roc_auc'], verbose=1, refit='f1') #verbose показывает время
grid_model.fit(features_trainplus, target_trainplus)

```

```python
print ("Метрика F1 на учебной выборке=", grid_model.best_score_)

results = pd.DataFrame(grid_model.cv_results_)
print('ROC AUC', results[results['mean_test_f1']== grid_model.best_score_]['mean_test_roc_auc'])
```
